## 1. Atoms, isotopes and relative atomic mass

Every atom has:
- a **proton number** $Z$ (also called atomic number) — the number of protons. Determines which element it is.
- a **mass number** $A$ — the number of protons + neutrons.
- so the **number of neutrons** is $A - Z$.

**Isotopes** are atoms of the same element with different mass numbers. The **relative atomic mass** ($A_r$) is the *weighted mean* of the isotope masses, weighted by their natural abundance.

$A_r = \dfrac{\sum (\text{mass} \times \text{abundance \%})}{100}$


### Worked example
Chlorine has two stable isotopes: $^{35}$Cl (75%) and $^{37}$Cl (25%). Find $A_r$.


````

# Weighted mean of the isotope masses.
A_r = (35 * 75 + 37 * 25) / 100
print(f"A_r(Cl) = {A_r}")   # 35.5 — matches the periodic table
````

## 2. Electron configuration and ions

Electrons sit in shells (energy levels). The simplest model fills shells in order: 2, 8, 8, 2 — i.e. shell 1 holds 2, shells 2 and 3 hold 8 each, then shell 4 starts with 2 (for K and Ca).

**Atoms become ions** by losing or gaining electrons to reach a full outer shell:
- Group 1 metals lose 1 electron → **+1** ion (e.g. Na⁺).
- Group 2 metals lose 2 electrons → **+2** ion (e.g. Mg²⁺).
- Group 7 non-metals gain 1 electron → **−1** ion (e.g. Cl⁻).
- Group 6 non-metals gain 2 electrons → **−2** ion (e.g. O²⁻).


### Worked example
Distribute 19 electrons of potassium across shells.


````

def shells(n_electrons):
    capacities = [2, 8, 8, 2]   # simple GCSE model
    out = []
    remaining = n_electrons
    for cap in capacities:
        if remaining <= 0:
            break
        out.append(min(cap, remaining))
        remaining -= cap
    return out

print("K (Z=19):", shells(19))   # [2, 8, 8, 1]
print("Ca (Z=20):", shells(20))  # [2, 8, 8, 2]
print("Cl (Z=17):", shells(17))  # [2, 8, 7]
````

## 3. Compounds, formulae and balancing equations

A chemical formula tells you which atoms are in a compound and in what ratio. We can parse a formula like `C6H12O6` into a `{'C': 6, 'H': 12, 'O': 6}` count of atoms.

**Balancing** an equation means making the count of each element equal on both sides — atoms are conserved in chemistry. You only ever add coefficients (the numbers in front), never change the formulae themselves.

$\mathrm{CH_4 + 2O_2 \rightarrow CO_2 + 2H_2O}$ — count both sides: 1 C, 4 H, 4 O on each.


### Worked example
Parse `H2SO4` into a dict of atom counts and check the formula mass.


````

import re

def parse_formula(formula):
    # Match an uppercase letter, optional lowercase letter, optional digits.
    pattern = re.compile(r'([A-Z][a-z]?)(\d*)')
    return {sym: int(count) if count else 1 for sym, count in pattern.findall(formula) if sym}

print(parse_formula('H2SO4'))     # {'H': 2, 'S': 1, 'O': 4}
print(parse_formula('CO2'))        # {'C': 1, 'O': 2}
print(parse_formula('C6H12O6'))    # {'C': 6, 'H': 12, 'O': 6}

# Relative formula mass M_r = sum of (count * A_r) for each element.
counts = parse_formula('H2SO4')
M_r = sum(n * A_R[el] for el, n in counts.items())
print(f"M_r(H2SO4) = {M_r}")   # 2*1 + 32 + 4*16 = 98
````

## 4. The mole and Avogadro's number

The **mole** is just a counting unit: 1 mol of anything is $6.022 \times 10^{23}$ particles — Avogadro's number, $N_A$.

Two formulae you'll use constantly:

$$n = \dfrac{m}{M_r}\quad\text{(moles from mass)}\qquad N = n \cdot N_A\quad\text{(particles from moles)}$$

where $n$ = moles, $m$ = mass in grams, $M_r$ = relative formula mass.

How many moles in 18 g of water? How many molecules?


````

M_r_water = 18                       # H2O: 2*1 + 16
m = 18                                # grams
n = m / M_r_water                     # moles
N = n * AVOGADRO                      # number of molecules

print(f"n = {n} mol")
print(f"N = {N:.3e} molecules")
````

## 5. Reacting masses and limiting reagents

Once an equation is balanced, the coefficients give the **mole ratio** of the reactants and products. To predict masses:

1. Convert masses to moles using $n = m/M_r$.
2. Use the mole ratio to find moles of the desired species.
3. Convert back to mass using $m = n \cdot M_r$.

When you have two reactants, the **limiting reagent** is the one that runs out first — the smaller of (mol available) ÷ (mol needed per equation).


### Worked example
How many grams of CO₂ form when 16 g of CH₄ burn? CH₄ + 2O₂ → CO₂ + 2H₂O.


````

m_CH4 = 16
M_r_CH4 = 16          # 12 + 4
M_r_CO2 = 44          # 12 + 32

n_CH4 = m_CH4 / M_r_CH4         # = 1.0 mol
n_CO2 = n_CH4 * (1 / 1)          # 1:1 mole ratio CH4 → CO2
m_CO2 = n_CO2 * M_r_CO2

print(f"n(CH4)  = {n_CH4} mol")
print(f"n(CO2)  = {n_CO2} mol")
print(f"m(CO2)  = {m_CO2} g")
````

## 6. Concentrations and dilutions

Concentration is moles (or mass) per unit volume:

$\text{conc (mol/dm}^3\text{)} = \dfrac{n}{V}$ where V is in **dm³** (= litres).
$\text{conc (g/dm}^3\text{)} = \dfrac{m}{V}$

**Dilution rule** $C_1 V_1 = C_2 V_2$ — the moles before and after dilution are conserved. Watch units: V can be in cm³, dm³, or m³ as long as both sides match.

Conversion: 1 dm³ = 1000 cm³ = 1 litre.


### Worked example
What concentration of NaOH (M_r=40) results from dissolving 8 g in 250 cm³ of water?


````

m = 8                # g
M_r = 40             # NaOH
V_cm3 = 250
V_dm3 = V_cm3 / 1000

n = m / M_r           # moles
conc = n / V_dm3      # mol/dm^3

print(f"n        = {n} mol")
print(f"conc     = {conc} mol/dm^3")
````

## 7. Acids, bases and pH

**pH** measures hydrogen-ion concentration on a log scale:

$\text{pH} = -\log_{10}[\mathrm{H}^+]\qquad [\mathrm{H}^+] = 10^{-\text{pH}}$

- pH < 7 → acidic
- pH = 7 → neutral (at 25 °C)
- pH > 7 → alkaline

A **strong acid** (e.g. HCl) fully dissociates — [H⁺] equals the acid concentration. A **weak acid** (e.g. ethanoic acid) only partially dissociates, so [H⁺] is less.

**Neutralisation**: acid + base → salt + water. With a strong monoprotic acid and monoprotic base, $n_\text{acid} = n_\text{base}$ at the end-point.

## 8. Energy changes

**Exothermic** reactions release energy (ΔH < 0); **endothermic** reactions absorb it (ΔH > 0).

**Bond enthalpy method**: every bond stores energy. To estimate the enthalpy change of a reaction:

$\Delta H \approx \sum E(\text{bonds broken}) - \sum E(\text{bonds formed})$

If more energy is needed to break bonds than is released forming new ones, ΔH is positive (endothermic) — and vice versa.


### Worked example
Estimate ΔH for $\mathrm{H_2 + Cl_2 \rightarrow 2HCl}$ given E(H–H)=436, E(Cl–Cl)=243, E(H–Cl)=432 kJ/mol.


````

broken = 436 + 243          # H-H + Cl-Cl
formed = 2 * 432            # 2 × H-Cl
delta_H = broken - formed   # kJ/mol

print(f"Bonds broken: {broken} kJ")
print(f"Bonds formed: {formed} kJ")
print(f"ΔH ≈ {delta_H} kJ/mol")
print("Negative → exothermic" if delta_H < 0 else "Positive → endothermic")
````

## 9. Rates of reaction

**Rate of reaction** = how fast a reactant is consumed (or product formed) per unit time.

$\text{rate} = \dfrac{\Delta[\text{reactant}]}{\Delta t}$

Factors that increase rate:
- Higher concentration → more particles per unit volume → more collisions.
- Higher temperature → more particles have ≥ activation energy.
- Smaller particle size → more surface area exposed.
- Catalyst → lowers activation energy without being consumed.

**Collision theory** (qualitative): a successful reaction needs particles to collide *with enough energy* and in the *right orientation*.

## 11. Chemical bonding

Three main bond types:

- **Ionic** — between a metal and a non-metal. Electrons are *transferred*. Forms a giant lattice. High melting points; conducts when molten or dissolved.
- **Covalent** — between two non-metals. Electrons are *shared*. Forms simple molecules (low melting points) or giant covalent structures (e.g. diamond, graphite, SiO₂ — very high melting points).
- **Metallic** — between metal atoms. A *sea of delocalised electrons* held by positive metal cations. Conducts; malleable.

A simple test: is at least one element a metal?


### Worked example
Classify NaCl, H₂O, CH₄, MgO, Cu, Fe.


````

METALS = {'Li','Na','K','Be','Mg','Ca','Al','Fe','Cu','Zn','Ag','Au','Pb'}

def bond_type(els):
    has_metal = any(e in METALS for e in els)
    has_nonmetal = any(e not in METALS for e in els)
    if has_metal and has_nonmetal: return 'ionic'
    if has_metal: return 'metallic'
    return 'covalent'

print(bond_type(['Na', 'Cl']))   # ionic
print(bond_type(['H', 'O']))      # covalent (H2O)
print(bond_type(['C', 'H']))      # covalent
print(bond_type(['Mg', 'O']))     # ionic
print(bond_type(['Cu']))           # metallic
````

## 12. Organic chemistry — alkanes, alkenes, alcohols

**Homologous series** are families of organic molecules sharing a general formula and similar properties:

| Series | General formula | Suffix | First member |
|---|---|---|---|
| Alkanes | $C_nH_{2n+2}$ | -ane | $\text{CH}_4$ (methane) |

Alkenes have a C=C double bond — the source of most of their reactivity. Alcohols carry an -OH group; their boiling points are much higher than the corresponding alkane's because of hydrogen bonding.
