## 1. Integers, negatives and BIDMAS

**BIDMAS** tells us the order to do calculations:
- **B**rackets
- **I**ndices (powers)
- **D**ivision and **M**ultiplication (left to right)
- **A**ddition and **S**ubtraction (left to right)

**Negatives:**
- Two negatives multiplied (or divided) give a positive: `-3 √ó -4 = 12`
- Adding a negative is the same as subtracting: `5 + (-3) = 2`
- Subtracting a negative is the same as adding: `5 - (-3) = 8`


### Worked example
Evaluate $2 + 3 \times (4 - 1)^2$

````python
# Step by step, following BIDMAS:
brackets = 4 - 1          # 3
indices  = brackets ** 2  # 9
multiply = 3 * indices    # 27
result   = 2 + multiply   # 29

print(f"Brackets:  4 - 1       = {brackets}")
print(f"Indices:   3^2         = {indices}")
print(f"Multiply:  3 √ó 9       = {multiply}")
print(f"Add:       2 + 27      = {result}")

# Python follows BIDMAS automatically, so this gives the same thing:
print("One-line:", 2 + 3 * (4 - 1)**2)
````

## 2. Fractions, decimals and percentages

These are three ways to write the same idea: *part of a whole*.

$$\frac{1}{2} = 0.5 = 50\%$$

**Adding fractions** ‚ÄĒ use a common denominator:
$$\frac{1}{3} + \frac{1}{4} = \frac{4}{12} + \frac{3}{12} = \frac{7}{12}$$

**Percentage of a number:** $X\%$ of $N$ is $\frac{X}{100} \times N$.

**Percentage change:** $\frac{\text{new} - \text{old}}{\text{old}} \times 100\%$.

In Python, the `Fraction` class handles fractions exactly (no rounding errors).


### Worked example
A jumper costs ¬£40 and is reduced by 25%. What's the new price?

````python
original = 40
discount_percent = 25

saving = (discount_percent / 100) * original
new_price = original - saving

print(f"Saving: ¬£{saving}")
print(f"New price: ¬£{new_price}")

# Or in one line, using a "multiplier":
multiplier = 1 - discount_percent / 100    # 0.75
print(f"One-liner: ¬£{original * multiplier}")
````

## 3. Ratio and proportion

A **ratio** compares two quantities: $a : b$.

To **share** $N$ in ratio $a : b$:
- Total shares $= a + b$
- One share $= N / (a+b)$
- Amounts: $a \times$ one share, and $b \times$ one share.

**Direct proportion:** $y = kx$. Double $x$, double $y$.

**Inverse proportion:** $y = k/x$. Double $x$, halve $y$. (Think: more workers ‚Üí less time.)


### Worked example
Share ¬£60 in the ratio 2 : 3.

````python
total = 60
a, b = 2, 3

total_shares = a + b      # 5
one_share = total / total_shares   # 12

amount_a = a * one_share  # 24
amount_b = b * one_share  # 36

print(f"Person A gets ¬£{amount_a}")
print(f"Person B gets ¬£{amount_b}")
print(f"Check: {amount_a} + {amount_b} = {amount_a + amount_b}")
````

## 4. Powers, roots and standard form

$a^n$ means multiply $a$ by itself $n$ times. $2^3 = 2 \times 2 \times 2 = 8$.

**Laws of indices:**
- $a^m \times a^n = a^{m+n}$
- $a^m \div a^n = a^{m-n}$
- $(a^m)^n = a^{mn}$
- $a^0 = 1$ and $a^{-n} = 1/a^n$

**Square root:** $\sqrt{x}$ is the number that squared gives $x$.

**Standard form:** $a \times 10^n$ where $1 \le a < 10$.
- $3{,}400{,}000 = 3.4 \times 10^6$
- $0.00072 = 7.2 \times 10^{-4}$


### Worked example
Simplify $2^3 \times 2^5$ two ways.

````python
# Way 1: compute directly
direct = 2**3 * 2**5

# Way 2: use the law  a^m √ó a^n = a^(m+n)
using_law = 2**(3 + 5)

print(f"Direct:   2^3 √ó 2^5 = {2**3} √ó {2**5} = {direct}")
print(f"Law:      2^(3+5)   = 2^8       = {using_law}")
````

## 5. Algebra: expressions and equations

A **variable** (like $x$) stands for an unknown number.

**To solve** $3x + 2 = 11$:
- Subtract 2 from both sides: $3x = 9$
- Divide both sides by 3: $x = 3$

**Expanding brackets:** $a(b + c) = ab + ac$

**Collecting like terms:** $4x + 3 - x + 1 = 3x + 4$

## 6. Sequences

An **arithmetic (linear) sequence** has a constant difference $d$:
$$\text{nth term} = a + (n-1)d$$

For $2, 5, 8, 11, \ldots$: $a=2, d=3$, so the nth term is $3n - 1$.

A **geometric sequence** has a constant ratio $r$: $\text{nth term} = ar^{n-1}$.

A **quadratic sequence** has constant second differences, e.g. $1, 4, 9, 16, \ldots$ (square numbers).


### Worked example
First 8 terms of an arithmetic sequence starting at 2, step 3.

````python
def arithmetic(a, d, num_terms):
    return [a + i * d for i in range(num_terms)]

terms = arithmetic(2, 3, 8)
print(terms)
# nth term formula: 2 + (n-1)*3 = 3n - 1
# 10th term: 3*10 - 1 = 29
print("10th term:", 2 + 9 * 3)
````

**Parallel** lines have equal gradients. **Perpendicular** lines: $m_1 \times m_2 = -1$.

## 10. Pythagoras and trigonometry

In a right-angled triangle with hypotenuse $c$ (longest side, opposite the right angle):
$$a^2 + b^2 = c^2 \quad \Rightarrow \quad c = \sqrt{a^2 + b^2}$$

**SOHCAHTOA** ‚ÄĒ for an angle $\theta$:
- $\sin\theta = \dfrac{\text{opposite}}{\text{hypotenuse}}$
- $\cos\theta = \dfrac{\text{adjacent}}{\text{hypotenuse}}$
- $\tan\theta = \dfrac{\text{opposite}}{\text{adjacent}}$

‚ö†ÔłŹ **Python trig uses radians, not degrees.** Convert with `math.radians(degrees)` and `math.degrees(radians)`.


### Worked example

## 11. Statistics

Given a list of numbers:
- **Mean** ‚ÄĒ add them all, divide by how many.
- **Median** ‚ÄĒ middle value when sorted (average of the two middle if even count).
- **Mode** ‚ÄĒ most frequent value.
- **Range** ‚ÄĒ largest minus smallest.

These are **measures of central tendency** and **spread**.


### Worked example
For the data `[4, 2, 8, 6, 5]`, compute all four.

## 12. Probability (with simulation)

Probability of an event $=\dfrac{\text{favourable outcomes}}{\text{total outcomes}}$.

Values lie between 0 (impossible) and 1 (certain).

- $P(\text{not } A) = 1 - P(A)$
- **Independent** events: $P(A \text{ and } B) = P(A) \times P(B)$
- **Mutually exclusive**: $P(A \text{ or } B) = P(A) + P(B)$

A fun thing about programming: instead of *calculating* probabilities, we can **simulate** experiments thousands of times and count what happens. The more simulations, the closer we get to the true probability (**the law of large numbers**).


### Worked example
What's the probability of getting at least one six in two rolls of a fair die? Compute it two ways: the formula, then a simulation.

````python
# Formula way:
# P(at least one 6) = 1 - P(no 6 at all) = 1 - (5/6)^2
p_exact = 1 - (5/6)**2
print(f"Exact probability: {p_exact:.4f}")

# Simulation way:
random.seed(42)   # makes results reproducible
trials = 100_000
successes = 0
for _ in range(trials):
    roll1 = random.randint(1, 6)
    roll2 = random.randint(1, 6)
    if roll1 == 6 or roll2 == 6:
        successes += 1

p_sim = successes / trials
print(f"Simulated:        {p_sim:.4f}")
print(f"Difference:       {abs(p_exact - p_sim):.4f}")
````

## 13. Quadratics

A **quadratic** is anything of the form $ax^2 + bx + c$. The graph is a parabola.

Three ways to solve $ax^2 + bx + c = 0$:
1. **Factorising**: rewrite as $(x - p)(x - q) = 0$ ‚ÄĒ solutions are $p, q$.
2. **Completing the square**: rewrite as $(x + p)^2 + q$ ‚ÄĒ useful for finding the vertex.
3. **Quadratic formula**: $x = \dfrac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ ‚ÄĒ always works.

The **discriminant** $\Delta = b^2 - 4ac$ tells you how many real solutions exist:
$\Delta > 0$ ‚áí two solutions, $\Delta = 0$ ‚áí one solution, $\Delta < 0$ ‚áí no real solutions.


### Worked example
Solve $x^2 - 5x + 6 = 0$ three ways.


````python

import math

# 1) Factorising: x^2 - 5x + 6 = (x - 2)(x - 3) = 0  ‚Üí  x = 2 or x = 3
print("Factored:", [2, 3])

# 2) Completing the square: x^2 - 5x + 6 = (x - 2.5)^2 - 0.25
#    (x - 2.5)^2 = 0.25  ‚Üí  x - 2.5 = ¬Ī0.5  ‚Üí  x = 2 or 3
p = -5 / 2
q = 6 - p**2
print(f"Completed:  (x + {p})^2 + {q}  ‚Üí  x = {-p + 0.5} or {-p - 0.5}")

# 3) Quadratic formula.
a, b, c = 1, -5, 6
disc = b**2 - 4*a*c
x1 = (-b + math.sqrt(disc)) / (2*a)
x2 = (-b - math.sqrt(disc)) / (2*a)
print(f"Formula:   discriminant={disc}, x = {x1}, {x2}")
````

## 15. Surds and advanced indices

**Surds** are irrational roots written exactly: $\sqrt{2}$, $\sqrt{3}$, $\sqrt{12}$ etc.

Useful identities:
- $\sqrt{ab} = \sqrt{a}\sqrt{b}$  ‚Üí simplify $\sqrt{12} = \sqrt{4 \cdot 3} = 2\sqrt{3}$.

**Index laws** you should know cold:
- $a^m \cdot a^n = a^{m+n}$,  $a^m / a^n = a^{m-n}$,  $(a^m)^n = a^{mn}$

## 16. Functions and rearranging formulae

**Function notation**: $f(x) = 3x + 2$ means "the function f sends x to 3x+2".
- $f(4) = 14$ ‚ÄĒ substitute the number for x.
- **Composite**: $fg(x) = f(g(x))$ ‚ÄĒ apply g first, then f.
- **Inverse** $f^{-1}$ undoes f. To find it, swap x and y, then solve for y.

**Rearranging formulae** (changing the subject) uses the same rules as solving equations: do the same to both sides until the variable you want is alone.

## 17. Advanced trigonometry

Right-angled trig (SOH-CAH-TOA) only works when there's a right angle. For *any* triangle, you have:

- **Sine rule**: $\dfrac{a}{\sin A} = \dfrac{b}{\sin B} = \dfrac{c}{\sin C}$
- **Cosine rule**: $a^2 = b^2 + c^2 - 2bc\cos A$
- **Area**: $\text{Area} = \dfrac{1}{2}ab \sin C$

Use sine rule when you have a *side and its opposite angle*. Use cosine rule when you have *all three sides* (to find an angle), or *two sides and the angle between them* (to find the third side). Python's `math.sin/cos` expect **radians** ‚ÄĒ convert with `math.radians`.


### Worked example
Triangle: a=7, b=10, angle C=42¬į. Find side c (cosine rule) and area.


````python

import math

a, b, C_deg = 7, 10, 42
C = math.radians(C_deg)

# Cosine rule for the third side: c^2 = a^2 + b^2 - 2ab*cos(C).
c = math.sqrt(a**2 + b**2 - 2*a*b*math.cos(C))
print(f"c = {c:.4f}")

# Area = 1/2 * a * b * sin(C).
area = 0.5 * a * b * math.sin(C)
print(f"area = {area:.4f}")
````

## 18. Vectors and GCSE statistics

**Vectors** in 2D have a direction and a magnitude. Written as a column $\binom{a}{b}$ or as $a\mathbf{i} + b\mathbf{j}$.
- Adding vectors: add the components.
- Scalar multiplication: multiply each component.
- Magnitude: $|\mathbf{v}| = \sqrt{a^2 + b^2}$ (Pythagoras).

**Cumulative frequency**: running total of the frequencies. The median sits at the n/2-th data point; the lower and upper quartiles at n/4 and 3n/4.

**Box plots** show min, lower quartile, median, upper quartile, max ‚ÄĒ the **five-number summary**. The **interquartile range** (IQR) is upper ‚ąí lower.


### Worked example
Vector arithmetic and a five-number summary on a small dataset.


````python

import math
import numpy as np

# Vector addition + magnitude.
v = (3, 4)
w = (1, 2)
v_plus_w = (v[0] + w[0], v[1] + w[1])
mag = math.sqrt(v[0]**2 + v[1]**2)
print(f"v + w = {v_plus_w}")
print(f"|v|   = {mag}")

# Five-number summary on a tiny dataset.
data = [5, 7, 8, 12, 13, 14, 18, 21, 22, 30]
q1, med, q3 = np.percentile(data, [25, 50, 75])
print(f"min={min(data)}  Q1={q1}  median={med}  Q3={q3}  max={max(data)}")
print(f"IQR = {q3 - q1}")
````
